```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint


a=1
c=1
b=0.03
d=0.02
N=500
t_f=20
ts=np.linspace(0,t_f,N)

y0=[10.,50.]


def volterra(y,t):
  y_1,y_2 = y
  dydt = [a*y_1-b*y_1*y_2,-c*y_2+d*y_1*y_2]
  return dydt



sol = odeint(volterra, y0, ts,full_output=True)


hares = (sol[0])[:,0]
foxes = (sol[0])[:,1]


sol[1].get('nst')

print('number of time steps for t_f = 20:',(sol[1].get('nst'))[N-2])


plt.plot(ts,hares)
plt.plot(ts,foxes)
plt.xlabel('time')
plt.legend(['hares','foxes'])
plt.ylabel('population')
plt.title('predator-prey populations against time')
plt.show()

legend = []
y_0s = [[10,100],[20,90],[30,70],[40,60],[50,50]]
for y_0 in y_0s:
  sol = odeint(volterra, y_0, ts,full_output=True)


  hares = (sol[0])[:,0]
  foxes = (sol[0])[:,1]


  legend.append(f'initial: hares ={y_0[0]}, foxes = {y_0[1]}' )
  plt.plot(hares,foxes)




plt.legend(legend)

plt.title('predator-prey populations against initial conditions')
plt.xlabel('hares')
plt.ylabel('foxes')
plt.show()




```
